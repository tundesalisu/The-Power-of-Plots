## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)



```

```python
mouse_metadata
```

```python
study_results
```

```python
# Combine the data into a single dataset
combined_data = pd.merge(mouse_metadata,study_results,on="Mouse ID", how="outer")
combined_data
```

```python
# Display the data table for preview
combined_data.head()
```

```python
# Checking the number of mice.
len(combined_data["Mouse ID"].unique())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

duplicated_combined_data = combined_data[combined_data.duplicated(["Mouse ID","Timepoint"])]
duplicated_combined_data.head(20)
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_combined_data = combined_data.loc[(combined_data["Mouse ID"] !="g989")]
clean_combined_data    
                                        
```

```python
# Checking the number of mice in the clean DataFrame.

len(clean_combined_data["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method is the most straighforward, creating multiple series and putting them all together at the end.
```

```python
clean_combined_data1 = clean_combined_data
# mean regimen drug
summary_clean_combined_data = clean_combined_data1.groupby(["Drug Regimen","Timepoint"])
mean_summary_clean_combined_data =  summary_clean_combined_data["Tumor Volume (mm3)"].mean()
mean_summary_clean_combined_data.head()
```

```python
#median summary 
median_summary_clean_combined_data =  summary_clean_combined_data["Tumor Volume (mm3)"].median()
median_summary_clean_combined_data
```

```python
# standard deviation
std_summary_clean_combined_data =  summary_clean_combined_data["Tumor Volume (mm3)"].std()
std_summary_clean_combined_data
```

```python
# Variance which is square of STD
variance_summary_clean_combined_data =  (summary_clean_combined_data["Tumor Volume (mm3)"].std())*(summary_clean_combined_data["Tumor Volume (mm3)"].std())
variance_summary_clean_combined_data
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method produces everything in a single groupby function
summary_statistics = pd.DataFrame({
    "mean" :mean_summary_clean_combined_data,
    "median":median_summary_clean_combined_data,
    "variance":variance_summary_clean_combined_data,
    "standard_deviation":std_summary_clean_combined_data
})
summary_statistics
```

## Correlation and Regression

```python

# for mouse weight and average tumor volume for the Capomulin regimen
```

```python
#and linear regression model 

mouse_wgt = clean_combined_data5_mean_Capomulin_result.iloc[:,1]
mean_tumor_vl = clean_combined_data5_mean_Capomulin_result.iloc[:,2]
correlation = st.pearsonr(mouse_wgt,mean_tumor_vl)
print(f"The correlation between both factors is {round(correlation[0],2)}")
```

```python
from scipy.stats import linregress
x_values = clean_combined_data5_mean_Capomulin_result["Weight (g)"]
y_values = clean_combined_data5_mean_Capomulin_result["Tumor Volume (mm3)"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values, y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel("Mouse Weight")
plt.ylabel("Mean Tumor Volume")
plt.show()
```
